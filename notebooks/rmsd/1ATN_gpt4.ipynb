```python
from mdagent import MDAgent
import time

myagent = MDAgent(
    agent_type='Structured',
    #model='gpt-3.5-turbo-0125',
    model='gpt-4-1106-preview',
    tools_model='gpt-3.5-turbo-0125',
    subagents_model='gpt-3.5-turbo-0125',
)

start_time = time.time()
myagent.run("Simulate protein 1ATN and calculate RMSD.")
end_time = time.time()
print("Time taken (s): ", end_time - start_time)
print("Time taken (min): ", (end_time - start_time)/60)
```
