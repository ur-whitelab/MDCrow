```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
models_unique = ['GPT-4o-2024-08-06', 'GPT-4-turbo-04-22-2024', 'GPT-3.5-turbo-0125']
descriptive_steps = [1.35, 1.34, 1.61]
less_descriptive_steps = [1.28, 1.58, 0.98]
descriptive_steps_success = [1.34, 1.24, 2.425]
less_descriptive_steps_success = [1.28, 1.40, 1.33]

# Create a DataFrame for the data
df = pd.DataFrame({
    'Model': np.tile(models_unique, 4),
    'Steps Ratio': descriptive_steps + less_descriptive_steps + descriptive_steps_success + less_descriptive_steps_success,
    'Description Type': ['Descriptive'] * 3 + ['Less Descriptive'] * 3 + ['Descriptive'] * 3 + ['Less Descriptive'] * 3,
    'Experiment Type': ['All Experiments'] * 6 + ['Successful Experiments'] * 6
})

# Set up the catplot
g = sns.catplot(
    data=df,
    x='Model', y='Steps Ratio', hue='Description Type', kind='bar',
    col='Experiment Type', height=6, aspect=1.2
)

# Add a horizontal line at y=1
for ax in g.axes.flat:
    ax.axhline(y=1, color='gray', linestyle='--')

# Adjust the title and layout
g.set_titles("{col_name}")
g.fig.subplots_adjust(top=0.85)
# g.fig.suptitle('Attempted Steps by Model and Type Conditioned on Experiment Success')

# Show plot
plt.show()
```
