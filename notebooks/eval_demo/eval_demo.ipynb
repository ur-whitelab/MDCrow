```python
from mdagent import Evaluator

evaluator = Evaluator(base_dir="evaluation_results") # that's the default - change base_dir if desired

# to do it manually
agent_params = {
    "agent_type": "Structured",
    "learn": False,
}
prompts = [
    'Download and clean fibronectin.',
    'Download and clean 1A3N.',
]
evaluator.run_and_evaluate(prompts, agent_params=agent_params)
evaluator.save()
```
